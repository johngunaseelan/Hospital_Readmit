# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
print(clean_hospital_read_df.loc[clean_hospital_read_df['Number of Discharges']<100, 'Excess Readmission Ratio'].mean())

print(clean_hospital_read_df.loc[(clean_hospital_read_df['Number of Discharges']<100) & (clean_hospital_read_df['Excess Readmission Ratio']>1.0), 'Excess Readmission Ratio'].count()/ clean_hospital_read_df.loc[clean_hospital_read_df['Number of Discharges']<100, 'Excess Readmission Ratio'].count())

print(clean_hospital_read_df.loc[clean_hospital_read_df['Number of Discharges']>1000, 'Excess Readmission Ratio'].mean())


print(clean_hospital_read_df.loc[(clean_hospital_read_df['Number of Discharges']>1000) & (clean_hospital_read_df['Excess Readmission Ratio']>1.0), 'Excess Readmission Ratio'].count()/ clean_hospital_read_df.loc[clean_hospital_read_df['Number of Discharges']>1000, 'Excess Readmission Ratio'].count())
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)


fig.tight_layout()
```

```python
df = clean_hospital_read_df

excess_radmin_ratio_lt100 = df.loc[(df['Number of Discharges']<100) & (df['Excess Readmission Ratio'].notnull()), 'Excess Readmission Ratio'].values
excess_radmin_ratio_gt1000 = df.loc[(df['Number of Discharges']>1000) & (df['Excess Readmission Ratio'].notnull()), 'Excess Readmission Ratio'].values
```

### Two sample z-test 

As we know the population mean and std we go for z test.

Null hypothesis : The mean excess readmission ratio is same for hospitals with less than 100 discharges and hospitals with greater than 1000 discharges.

Alternate hypothesis : The mean excess readmission ratio is not same for hospitals with less than 100 discharges and hospitals with greater than 1000 discharges.

Given alpha : 0.01

### Since p-value(8.24e-24) is less than the alpha(0.01) value the null hypothesis is rejected. The mean excess readmission rate is different for hospitals with less than 100 discharges and hospitals greater than 1000 discharges. This is practically significant as p value is much less than the given alpha of 0.01. 

```python
z_score, p_value = ztest(excess_radmin_ratio_lt100, excess_radmin_ratio_gt1000)

print('z score : ', z_score)
print('p value : ', p_value)
```
